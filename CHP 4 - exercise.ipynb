```python
import matplotlib.pyplot as plt
from sklearn import datasets
digits = datasets.load_digits()
```

```python
fig, ax = plt.subplots(5, 10, figsize=(8, 4), subplot_kw={'xticks': [], 'yticks': []}, constrained_layout=True)

for i in range(digits.data[:50].shape[0]):
  row = i // 10
  col = i % 10
  ax[row, col].imshow(digits.data[i].reshape(8, 8), cmap='gray')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state=42, test_size=0.25, stratify=digits.target)
print(f'X_train_shape: {X_train.shape}')
print(f'X_test_shape: {X_test.shape}')
print(f'y_train_shape: {y_train.shape}')
print(f'y_test_shape: {y_test.shape}')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN

knn = KNN(n_neighbors=3)
knn.fit(X_train_scaled, y_train)

accuracy = knn.score(X_test_scaled, y_test)
print(f'The accuracy is: {accuracy}')
```
